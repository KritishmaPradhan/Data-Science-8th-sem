**T test example pgm for sales data**

```python
from scipy import stats
import numpy as np
# group A Sales without a discount(per day data)
group_a = [200,210,190,205,195]
# group B Sales with a discount
group_b = [250, 240,260,230,255]

t_stat, p_val = stats.ttest_ind(group_a, group_b)
print(f"p_value : {p_val:.4f}")
if p_val < 0.05:
    print("The discount significantly increased sales!")
else:
    print("The difference might just be luck")

```

```python
# T test example pgm for teaching data
from scipy import stats
import numpy as np
# group A: Scores of student using Trditional Boooks
group_a = [60,70,62,68,72]
# group B Scores of students using Interactive AI Tools
group_b = [85,88,90,82,87]

t_stat, p_val = stats.ttest_ind(group_a, group_b)
print(f"p_value : {p_val:.4f}")
if p_val < 0.05:      # 95% sure about AI tools improve learning outcome
    print("AI tools significantly improved learning outcomes!")
else:
    print("No significant difference between teaching methods")
```
