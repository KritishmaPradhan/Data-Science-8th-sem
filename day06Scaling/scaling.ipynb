Introduction to Scaling

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler,RobustScaler
#create a simple series (e.g.Price in Lakhs)
prices = pd.Series([10,12,15,14,13,50], name = "Price") 
#Scikit-learn expects a 2D array, so we reshape(-1,1)
#converts 1D data to 2D
# reshape (-1,1) converts an array into a column vector
data_2d = prices.values.reshape(-1,1)    #also try (1,-1)  -> (row,column)
print(data_2d)
```

```python
#initialize scalers
std_scalers = StandardScaler()
minmax_scaler = MinMaxScaler()
robust_scaler = RobustScaler()
```

```python
#create a comparison Dataframe 
results = pd.DataFrame({
    "Original" : prices,
    "Standardized (Z-score)" : std_scalers.fit_transform(data_2d).flatten(),
    "Normalized (0-1)" : minmax_scaler.fit_transform(data_2d).flatten(),
    "Robust (IQR)" : robust_scaler.fit_transform(data_2d).flatten()
})
print(results)
```
