***Multiple Regeression Implementation***

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
dataHouse = pd.DataFrame({
    'SquareFoot': [1500,2000,1200,2500,1800,1600,2200,1400,1900,2100,2800,1300,1700,2400,1100],
    'Bedrooms': [2,4,2,4,3,3,4,2,3,4,5,2,3,4,2],
    'HouseAge': [10,5,20,2,15,8,1,12,7,4,1,25,11,3,30],
    'Price': [300,450,200,550,350,320,480,280,380,420,600,180,340,500,150]
})
dataHouse
```

```python
plt.figure(figsize = (10,6))
sns.heatmap(dataHouse.corr(), annot = True)
plt.title("Correlation among house data")
plt.show()
```

```python
X = dataHouse[['SquareFoot', 'Bedrooms', 'HouseAge']]
Y = dataHouse['Price']
model = LinearRegression().fit(X,Y)
print(f"Intercept(c): {model.intercept_:.2f}")
print(f"Coefficients (mi): {model.coef_}")
```

```python
predicted_house_price = model.predict([[1800, 3, 12]])
print(f" the house price when sq ft is 1800 and bedroom is 3 and house age is 12 yrs is {predicted_house_price[0]:.2f}")
```
